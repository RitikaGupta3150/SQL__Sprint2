# Data Cleaning

```python
query = """
SELECT 
    COUNT(*) AS Total_Records,
    SUM(CASE WHEN Store_ID IS NULL THEN 1 ELSE 0 END) AS Missing_Store_ID,
    SUM(CASE WHEN Product_ID IS NULL THEN 1 ELSE 0 END) AS Missing_Product_ID,
    SUM(CASE WHEN Sale_Date IS NULL THEN 1 ELSE 0 END) AS Missing_Sale_Date,
    SUM(CASE WHEN Units_Sold IS NULL THEN 1 ELSE 0 END) AS Missing_Units_Sold,
    SUM(CASE WHEN Sales_Revenue_USD IS NULL THEN 1 ELSE 0 END) AS Missing_Sales_Revenue
FROM SalesData;
"""
df_missing_values = pd.read_sql(query, connection)
print(df_missing_values)
```

# Exploratory Data Analysis

```python
# SQL query for average sales revenue each day
query_avg_sales = """
SELECT Day_of_the_Week, AVG(Sales_Revenue_USD) AS Avg_Sales_Revenue
FROM SalesData
GROUP BY Day_of_the_Week;
"""

# Execute and fetch results
df_avg_sales = pd.read_sql(query_avg_sales, connection)
print(df_avg_sales)

# SQL query for average units sold each day
query_avg_units = """
SELECT Day_of_the_Week, AVG(Units_Sold) AS Avg_Units_Sold
FROM SalesData
GROUP BY Day_of_the_Week;
"""

# Execute and fetch results
df_avg_units = pd.read_sql(query_avg_units, connection)
print(df_avg_units)
```

Insight-calculates the average sales revenue and average units sold for each day of the week using groupby

```python

query_avg_sales = """
SELECT Day_of_the_Week, AVG(Sales_Revenue_USD) AS Avg_Sales_Revenue
FROM SalesData
GROUP BY Day_of_the_Week;
"""
df_avg_sales = pd.read_sql(query_avg_sales, connection)
plt.figure(figsize=(10, 6))
plt.bar(df_avg_sales['Day_of_the_Week'], df_avg_sales['Avg_Sales_Revenue'])
plt.xlabel('Day of the Week')
plt.ylabel('Average Sales Revenue (USD)')
plt.title('Average Sales Revenue by Day of the Week')
plt.show()
query_avg_units = """
SELECT Day_of_the_Week, AVG(Units_Sold) AS Avg_Units_Sold
FROM SalesData
GROUP BY Day_of_the_Week;
"""
df_avg_units = pd.read_sql(query_avg_units, connection)
plt.figure(figsize=(10, 6))
plt.bar(df_avg_units['Day_of_the_Week'], df_avg_units['Avg_Units_Sold'])
plt.xlabel('Day of the Week')
plt.ylabel('Average Units Sold')
plt.title('Average Units Sold by Day of the Week')
plt.show()
```

Insight-Visualizing daily averages highlights trends for revenue and sales, guiding strategic decisions.

```python
query_location_data = """
SELECT 
    Store_Location, 
    SUM(Sales_Revenue_USD) AS Total_Sales_Revenue,
    SUM(Units_Sold) AS Total_Units_Sold
FROM SalesData
GROUP BY Store_Location
ORDER BY Total_Sales_Revenue DESC, Total_Units_Sold DESC
LIMIT 10;
"""
df_location_data = pd.read_sql(query_location_data, connection)

print(df_location_data)
```

# Inferential Statistics

```python
query_discount_sales = """
SELECT Discount_Percentage, Sales_Revenue_USD
FROM SalesData;
"""
df_discount_sales = pd.read_sql(query_discount_sales, connection)

plt.figure(figsize=(10, 6))
plt.scatter(df_discount_sales['Discount_Percentage'], df_discount_sales['Sales_Revenue_USD'])
plt.xlabel('Discount Percentage')
plt.ylabel('Sales Revenue (USD)')
plt.title('Correlation between Discount Percentage and Sales Revenue')
plt.show()
```

Insight-The scatter plot helps identify trends or patterns, revealing how discounts impact sales revenue.

```python
query_sales_by_category = """
SELECT Product_Category, AVG(Sales_Revenue_USD) AS Avg_Sales_Revenue
FROM SalesData
GROUP BY Product_Category;
"""
df_sales_category = pd.read_sql(query_sales_by_category, connection)
df_sales_category.set_index('Product_Category').plot(kind='bar', legend=False)
plt.title('Most Products Sold by Category')
plt.xlabel('Product Category')
plt.ylabel('Average Revenue')
plt.show()
```

Insight-Analyzing product counts helps identify popular categories, guiding inventory and marketing strategies.

```python
query_location_sales = """
SELECT Store_Location, SUM(Sales_Revenue_USD) AS Total_Sales_Revenue
FROM SalesData
GROUP BY Store_Location
ORDER BY Total_Sales_Revenue DESC
LIMIT 10;  -- Top 10 locations
"""
df_location_sales = pd.read_sql(query_location_sales, connection)
plt.figure(figsize=(12, 6))
sns.barplot(data=df_location_sales, x='Store_Location', y='Total_Sales_Revenue')
plt.xticks(rotation=45, ha='right')
plt.title('Total Sales Revenue by Location (Top 10)', fontsize=14)
plt.xlabel('Store Location', fontsize=12)
plt.ylabel('Total Sales Revenue (USD)', fontsize=12)
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

Insight - Total Sales Revenue By Location
